### Example I: multiplication

- Multiplication can be regarded as adding itself with designate times

```python
# iterative thinking
def multiI(x, a):
    """
    x: multiplicand
    a: multiplicator
    """
    output = 0
    for i in range(a):
        output += x
    return output

# recursive thinking
def multiR(x, a):
    """
    x: multiplicand
    a: multiplicator
    """
    if a == 1:
        return x
    else:
        return x + multiR(x, a-1)
```

### Example II: factorial

```python
# iterative thinking
def factI(n):
    """
    Return n!
    """
    output = 1
    for i in range(n):
        output *= (i+1)
    return output

# recursive thinking
def factR(n):
    """
    Return n!
    """
    if n == 1:
        return n
    else:
        return n * factR(n-1)
```
